```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
from sklearn.datasets import load_diabetes
```

```python
df1 = load_diabetes()
df1
```

```python
print(df1['DESCR'])
```

```python
df = pd.DataFrame(df1.data,columns=['age',
                                   'sex',
                                   'bmi',
                                   'bp',
                                   's1',
                                   's2',
                                   's3',
                                   's4',
                                   's5',
                                   's6'])
```

```python
df.head()
```

```python
X=df
import numpy as np

# Assuming df1['target'] is your target variable
y = df1['target']  # y is a Pandas Series

# Check if y is a Pandas Series and convert to 2D array
if isinstance(y, pd.Series):
    y = y.values.reshape(-1, 1)  # Convert to 2D array
else:
    y = y.reshape(-1, 1)  # If y is already a NumPy array

# Now y should be in 2D shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=12)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),color='r')
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
import seaborn as sns

# Train the model
reg = DecisionTreeRegressor(max_depth=4)
reg.fit(X_train, y_train)

# Predict on the test set (use X_test and y_test)
y_pred = reg.predict(X_test)

# Plot predicted values against actual test values
plt.scatter(x=y_pred, y=y_test,color = 'g')
```
